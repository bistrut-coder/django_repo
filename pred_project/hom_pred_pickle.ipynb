```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_squared_error
```

```python
data = pd.read_csv(r"D:\Note's\DataScience_prakash _senapaty_10.30am\ML\SLR\26th- SLR\26th- SLR\SLR - House price prediction\House_data.csv")
```

```python
data.head()
```

```python
price = data["price"]
space = data["sqft_living"]
```

```python
x = np.array(price).reshape(-1,1)
y = np.array(space)
```

```python
x_train, x_test,y_train, y_test = train_test_split(x, y, random_state=0, test_size=0.20) 
```

```python
lr = LinearRegression()
lr.fit(x,y)
y_pred = lr.predict(x_test)
```

```python
plt.scatter(x_train, y_train, color = "red")
plt.plot(x_train, lr.predict(x_train), color = "blue")
plt.title("price vs space (train model)")
plt.xlabel("space")
plt.ylabel("price")
plt.show()
```

```python
plt.scatter(x_test, y_test, color = "red")
plt.plot(x_train, lr.predict(x_train), color = "blue")
plt.title("price vs space (train model)")
plt.xlabel("space")
plt.ylabel("price")
plt.show()
```

```python
bias = lr.score(x_train,y_train)
varience = lr.score(x_test,y_test)
print(f"The r2 of Train data is :{bias:,.2f}")
print(f"The r2 of Test data is : {varience:,.2f}")
```

```python
MSE_OF_train = mean_squared_error(y_train,lr.predict(x_train))
MSE_OF_test = mean_squared_error(y_test,y_pred)
```
